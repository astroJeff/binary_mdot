### First let's calculate the evolution of a binary through (super) simple N-body integration 

```python
# Calculate the instantaneous orbital separation and eccentricity

def A_instant(x1,y1,vx1,vy1,x2,y2,vx2,vy2,M1,M2):

    # distance between stars
    dist = np.sqrt((x2-x1)**2 + (y2-y1)**2)
    
    # center of mass velocity
    vcm_x = (M1*vx1 + M2*vx2) / (M1+M2)
    vcm_y = (M1*vy1 + M2*vy2) / (M1+M2)
    vcm = np.sqrt(vcm_x**2 + vcm_y**2)
    
    # adjust velocities based on center of mass
#    v1 = np.sqrt(vx1**2 + vy1**2)
#    v2 = np.sqrt(vx2**2 + vy2**2)
    v1 = np.sqrt((vx1 - vcm_x)**2 + (vy1 - vcm_y)**2)
    v2 = np.sqrt((vx2 - vcm_x)**2 + (vy2 - vcm_y)**2)
    
    # Calculate the orbital separation
    E_orb = (0.5*M1*v1**2) + (0.5*M2*v2**2) - G*M1*M2/dist
    A_inst = -G*M1*M2/(2.0*E_orb)
    
    # Find center of mass
    cm_x = (M1*x1 + M2*x2)/(M1+M2)
    cm_y = (M1*y1 + M2*y2)/(M1+M2)    
    
    # Move position of stars to center of mass
    x1 = x1 - cm_x
    x2 = x2 - cm_x
    y1 = y1 - cm_y
    y2 = y2 - cm_y
    
    # Now calculate angular momentum: sum(m*(vxR))
    J = M1*(x1*vy1 - y1*vx1) + M2*(x2*vy2 - y2*vx2)
    temp_1_e2 = (M1+M2)/((M1*M2)**2) * J**2 /(G*A_inst)
    e_inst = np.sqrt(1.0-temp_1_e2)
    
    return A_inst, e_inst
    
    
def A_to_P(A_in, M1, M2):
    return 2.0*np.pi * np.sqrt(A_in**3 / (G * (M1+M2)))

def P_to_A(P_in, M1, M2):
    return (G*(M1+M2) * (P_in/(2.0*np.pi))**2)**(1.0/3.0)
```

```python

# Constants all in cgs
Msun = 1.989e33 
G = 6.674e-8
secday = 3600.0*24.0
secyer = 3600.0*24.0*365.25
AUincm = 1.496e13 # AU in cm

# Set initial masses
M1_0 = 1.0 * Msun # in g
M2_0 = 1.0 * Msun # in g

# Initial positions
x1_0 = -1.0e3*AUincm *0.995# in cm
y1_0 = 0.0
x2_0 = 1.0e3*AUincm *0.995# in cm
y2_0 = 0.0
dist = np.sqrt((x1_0 - x2_0)**2 + (y1_0 - y2_0)**2)

# Initial velocities
vx1_0 = 0.0
vy1_0 = np.sqrt(G*M2_0/dist/2.0)
vx2_0 = 0.0
vy2_0 = -np.sqrt(G*M1_0/dist/2.0)

# Test for non-circular orbits
vy1_0 = vy1_0 * 0.1
vy2_0 = vy2_0 * 0.1

A_0, e_0 = A_instant(x1_0,y1_0,vx1_0,vy1_0,x2_0,y2_0,vx2_0,vy2_0,M1_0,M2_0)
P_0 = A_to_P(A_0,M1_0,M2_0)

# solve for initial eccentric anomaly
if e_0 == 0:
    E_0 = 0.0
    f_0 = 0.0
    e_0 = 1.0e-10
else:
    E_0 = np.arccos((1.0/e_0 * (1.0 - dist/A_0))+1.0e-8)
    f_0 = np.arccos(1.0/e_0* (A_0 / dist * (1.0-e_0*e_0) - 1.0)+1.0e-8)
        
print "Initial M1:", M1_0/Msun, "Msun, and M2:", M2_0/Msun, "Msun"
print "Initial orbital separation:", np.floor(10.0*A_0/AUincm)/10.0, "AU, and eccentricity:", e_0
print "Orbital period:", np.floor(10.0*P_0/secyer)/10.0, "years, or", np.floor(10.0*P_0/secday)/10.0, "days"
print "Initial eccentric anomaly:", E_0
print "Initial true anomaly:", f_0

# Set initial values
t0 = 0.0
z0 = [vx1_0, vy1_0, x1_0, y1_0, vx2_0, vy2_0, x2_0, y2_0, M1_0, M2_0]

# Set mass loss rates
Mdot_1 = -1.0e-7 * Msun / secyer
Mdot_2 = -1.0e-7 * Msun / secyer

# Set the end time and sample time for integration
t_max = 1.0 * P_0 # Integration time in seconds
dt = P_0 / 1000.0 # Take samples every thousandth of an orbit
```

### Integrate using N-Body

```python
# Define derivatives to integrate
def F(t,z,args):
    (v_x1, v_y1, x1, y1, v_x2, v_y2, x2, y2, M_1, M_2) = z
    (Mdot_1, Mdot_2) = args
    
    # Gravitational constant in cgs
    G = 6.674e-8
    
    # Distances
    delta_x = x2-x1
    delta_y = y2-y1
    d = np.sqrt(delta_x*delta_x + delta_y*delta_y)
    
    # Derivatives for 1st object
    dv_x1 = G*M_2*delta_x/(d**3)
    dv_y1 = G*M_2*delta_y/(d**3)
    dx1 = v_x1
    dy1 = v_y1
    
    # Derivatives for 2nd object
    dv_x2 = -G*M_1*delta_x/(d**3)
    dv_y2 = -G*M_1*delta_y/(d**3)
    dx2 = v_x2
    dy2 = v_y2

    # Mass loss - set to zero right now
    dM_1 = Mdot_1
    dM_2 = Mdot_2
    
#    print y1, y2

    return np.array([dv_x1, dv_y1, dx1, dy1, dv_x2, dv_y2, dx2, dy2, dM_1, dM_2])
    
```

```python
# Set up the integrator
r = ode(F).set_integrator('dop853', nsteps=1000, first_step=1.0, atol=1.0e-10, rtol=1.0e-10)

r.set_initial_value(z0, t0)
r.set_f_params([Mdot_1, Mdot_2])

# Run the integrator
time = np.array([])
x1 = np.array([])
y1 = np.array([])
x2 = np.array([])
y2 = np.array([])
vx1 = np.array([])
vy1 = np.array([])
vx2 = np.array([])
vy2 = np.array([])
M1 = np.array([])
M2 = np.array([])

while r.successful() and r.t < t_max:
    r.integrate(r.t+dt)
    
    time = np.append(time, r.t)
    x1 = np.append(x1, r.y[2])
    y1 = np.append(y1, r.y[3])
    x2 = np.append(x2, r.y[6])
    y2 = np.append(y2, r.y[7])
    vx1 = np.append(vx1, r.y[0])
    vy1 = np.append(vy1, r.y[1])
    vx2 = np.append(vx2, r.y[4])
    vy2 = np.append(vy2, r.y[5])
    M1 = np.append(M1, r.y[8])
    M2 = np.append(M2, r.y[9])
 
```

### Integrate using new equations

```python
def F_new(t,z,args):
    (A, e, E, M_1, M_2) = z
    (Mdot_1, Mdot_2) = args

    # Mean anomaly
    n = np.sqrt(G*(M_1+M_2)/A**3)
    mdot_m = (Mdot_1+Mdot_2) / (M_1+M_2)
    
    # Derivatives
    dA_dt = - (1.0+e*np.cos(E)) / (1.0-e*np.cos(E)) * mdot_m * A
    de_dt = - (1.0-e*e) * np.cos(E) / (1.0-e*np.cos(E)) * mdot_m
    dE_dt = 1.0 / (1.0-e*np.cos(E)) * (n - np.sin(E)*np.cos(E)*mdot_m)
    dM1_dt = Mdot_1
    dM2_dt = Mdot_2
    
    return np.array([dA_dt, de_dt, dE_dt, dM1_dt, dM2_dt])
    
```

```python
# Set up the integrator
r = ode(F_new).set_integrator('dop853', nsteps=1000, first_step=1.0, atol=1.0e-10)

z0 = [A_0, e_0, E_0, M1_0, M2_0]
r.set_initial_value(z0, t0)
r.set_f_params([Mdot_1, Mdot_2])

# Run the integrator
time_new = np.array([])
A_new = np.array([])
e_new = np.array([])
E_new = np.array([])
M1_new = np.array([])
M2_new = np.array([])

while r.successful() and r.t < t_max:
    r.integrate(r.t+dt)
    
    time_new = np.append(time_new, r.t)
    A_new = np.append(A_new, r.y[0])
    e_new = np.append(e_new, r.y[1])
    E_new = np.append(E_new, r.y[2])
    M1_new = np.append(M1_new, r.y[3])
    M2_new = np.append(M2_new, r.y[4])
```

### Integrate using true anomaly equations

```python
def F_true(t,z,args):
    (A, e, f, M_1, M_2) = z
    (Mdot_1, Mdot_2) = args

    # Mean anomaly
    n = np.sqrt(G*(M_1+M_2)/A**3)
    # Mdot over M
    mdot_m = (Mdot_1+Mdot_2) / (M_1+M_2)
    # 1-e^2
    e2_1 = 1.0-e*e
    
    # Derivatives
    dA_dt = - (1.0+2.0*e*np.cos(f) + e*e) / (1.0-e*e) * mdot_m * A
    de_dt = - (np.cos(f) + e) * mdot_m
    df_dt = np.sqrt(e2_1)/(e2_1*e2_1)*(1.0+e*np.cos(f))**2*n
    # Add precession term
    df_dt = df_dt + np.sin(f)/e * mdot_m
    
    dM1_dt = Mdot_1
    dM2_dt = Mdot_2
    
    return np.array([dA_dt, de_dt, df_dt, dM1_dt, dM2_dt])
    
```

```python
# Set up the integrator
r_true = ode(F_true).set_integrator('dop853', nsteps=1000, first_step=0.01, atol=1.0e-15)

z0 = [A_0, e_0, f_0, M1_0, M2_0]
r_true.set_initial_value(z0, t0)
r_true.set_f_params([Mdot_1, Mdot_2])

# Run the integrator
time_true = np.array([])
A_true = np.array([])
e_true = np.array([])
f_true = np.array([])
M1_true = np.array([])
M2_true = np.array([])

while r_true.successful() and r_true.t < t_max:
    r_true.integrate(r_true.t+dt)
    
    time_true = np.append(time_true, r_true.t)
    A_true = np.append(A_true, r_true.y[0])
    e_true = np.append(e_true, r_true.y[1])
    f_true = np.append(f_true, r_true.y[2])
    M1_true = np.append(M1_true, r_true.y[3])
    M2_true = np.append(M2_true, r_true.y[4])
```
