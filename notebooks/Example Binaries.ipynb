```python
import numpy as np
from scipy.integrate import ode
import matplotlib.pyplot as plt
from matplotlib import ticker
%matplotlib inline
```

### Equations for evolution of the orbital elements

```python
# Constants all in cgs
Msun = 1.989e33 
G = 6.674e-8
secday = 3600.0*24.0
secyer = 3600.0*24.0*365.25
AUincm = 1.496e13 # AU in cm

# Evolution using the true anomaly
def F_true(t,z,args):
    (A, e, f, M_1, M_2) = z
    (Mdot_1, Mdot_2) = args

    n = np.sqrt(G*(M_1+M_2)/A**3)    # Mean anomaly
    mdot_m = (Mdot_1+Mdot_2) / (M_1+M_2)    # Mdot over M
    e2_1 = 1.0-e*e    # 1-e^2
    
    # Derivatives
    dA_dt = - (1.0+2.0*e*np.cos(f) + e*e) / (1.0-e*e) * mdot_m * A
    de_dt = - (np.cos(f) + e) * mdot_m
    df_dt = np.sqrt(e2_1)/(e2_1*e2_1)*(1.0+e*np.cos(f))**2*n
    # Add precession term
    df_dt = df_dt + np.sin(f)/e * mdot_m
    
    dM1_dt = Mdot_1
    dM2_dt = Mdot_2
    
    return np.array([dA_dt, de_dt, df_dt, dM1_dt, dM2_dt])
   
# Kepler's 3rd law    
def A_to_P(A_in, M1, M2):
    return 2.0*np.pi * np.sqrt(A_in**3 / (G * (M1+M2)))

def P_to_A(P_in, M1, M2):
    return (G*(M1+M2) * (P_in/(2.0*np.pi))**2)**(1.0/3.0)    
```

### Set up integrator

```python
orbit = ode(F_true).set_integrator('dop853', nsteps=1000, first_step=1.0, atol=1.0e-10, rtol=1.0e-10)

# 1 Msun + 1 Msun binary
M1_0 = 1.0 * Msun
M2_0 = 1.0 * Msun
Mdot_1 = -1.0e-7 * Msun / secyer
Mdot_2 = -1.0e-7 * Msun / secyer
orbit.set_f_params([Mdot_1, Mdot_2])

t0 = 0.0
```

### Run integrator for 10 orbits

```python
#P_0 = 1.0e2 * secyer # yrs
e_0 = 0.99
#A_0 = P_to_A(P_0, M1_0, M2_0)
A_0 = 1000.0*AUincm
P_0 = A_to_P(A_0, M1_0, M2_0)
f_0 = 0.01
z0 = [A_0, e_0, f_0, M1_0, M2_0]
orbit.set_initial_value(z0, t0)

print "Initial eccentricity:", e_0
print "Initial semi-major axis:", A_0/AUincm, "AU"
print "Initial orbital period:,", P_0/secyer, "years"

# Take samples every hundredth of an orbit
dt = P_0 / 100.0 

# Integrate for 10 orbits
t_max = 10.0 * P_0

# Create the array to store orbital elements
A1000_e99 = np.array([t0, A_0, e_0, f_0, M1_0, M2_0])

# Run integrator
# Store data in array: [time, A, e, f, M1, M2]
while orbit.successful() and orbit.t < t_max:
    orbit.integrate(orbit.t+dt)
    A1000_e99 = np.vstack((A1000_e99, [orbit.t,orbit.y[0],orbit.y[1],orbit.y[2],orbit.y[3],orbit.y[4]]))
A1000_e99 = A1000_e99.T
```

### Plot 10 orbits

```python
f, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=False, figsize=(8,8))
ax1.plot(A1000_e99[0]/secyer/1000.0, A1000_e99[2] - 0.99)
ax1.set_title(r'$1.0\, M_{\odot} + 1.0\, M_{\odot}\, {\rm Binary},\, \dot{M} = -10^{-7}M_{\odot}/{\rm yr}$', fontsize=18, y=1.05)
#plt.title(figure_title, y=1.08)
ax1.set_ylabel(r'$\Delta$ Eccentricity', fontsize=16)
#ax1.set_ylim(0.989998,0.990002)
ax2.set_xlabel(r'Time ($10^3$ yrs)', fontsize=16)
ax2.plot(A1000_e99[0]/secyer/1000.0, A1000_e99[1]/AUincm)
ax2.set_ylabel('Semi-major Axis (AU)', fontsize=16)
#ax2.set_ylim(999.0, 1029.0)
# Fine-tune figure; make subplots close to each other and hide x ticks for
# all but bottom plot.

ax1.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0e'))
ax1.locator_params(nbins=5)
ax1.set_ylim(-8e-6,8e-6)

ax1.tick_params(axis='y', labelsize=14)


ax2.tick_params(axis='x', labelsize=14)
ax2.tick_params(axis='y', labelsize=14)
ax2.yaxis.set_major_formatter(ticker.ScalarFormatter(useOffset=False))
ax2.locator_params(nbins=7)




f.subplots_adjust(hspace=0)

f.savefig('../figures/1_1_binary_10orbits.pdf',bbox_inches='tight')

```

### Integrate for 200 orbits

```python


# Run for longer - 200 orbits
t0 = 0.0
t_max = 200.0 * P_0
orbit.set_initial_value(z0, t0)

# Remake the array to store orbital elements
A1000_e99 = np.array([t0, A_0, e_0, f_0, M1_0, M2_0])


# Run integrator
# Store data in array: [time, A, e, f, M1, M2]
while orbit.successful() and orbit.t < t_max:
    orbit.integrate(orbit.t+dt)
    A1000_e99 = np.vstack((A1000_e99, [orbit.t,orbit.y[0],orbit.y[1],orbit.y[2],orbit.y[3],orbit.y[4]]))
A1000_e99 = A1000_e99.T
```
